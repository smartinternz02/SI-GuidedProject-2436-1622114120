# Split The Data Into Independent And Dependent Variables

```python
x = data.iloc[:,0:8].values
y = data.iloc[:,8].values
```

```python
x
```

```python
y
```

# Split The Dataset Into Train Set And Test Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Training And Testing The Model

Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
dtpred = dt.predict(x_test)
```

```python
dtpred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
dtaccuracy=r2_score(y_test,dtpred)
```

```python
dtaccuracy
```

```python
dt.predict([['540.0','0.0','0.0','162.0','2.5','1040.0','676.0','28']])
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train,y_train)
```

```python
mlrp = mlr.predict(x_test)
```

```python
mlraccuracy = r2_score(y_test,mlrp)
```

```python
mlraccuracy
```

Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators =900)
rf.fit(x_train,y_train)
```

```python
rfpred =  rf.predict(x_test)
```

```python
rfaccuracy = r2_score(y_test,rfpred)
```

```python
rfaccuracy
```

```python
rf.predict([['540.0','0.0','0.0','162.0','2.5','1040.0','676.0','28']])
```

# Model Evaluation

```python
print("Accuracy of \nDecision Tree Regression = {0}\nMulti Linear Regression = {1}\nRandom Forest Regression = {2}\n"  .format(dtaccuracy,mlraccuracy,rfaccuracy))
```

```python
plt.scatter(y_test, mlrp)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Multi Linear Regrssion")
plt.show()

plt.scatter(y_test, dtpred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Decision Tree Regressor")
plt.show()

plt.scatter(y_test, rfpred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Random Forest Regressor")
plt.show()
```

by above accuracy and the graph, Random Forest Regressor is the best choice for the model
